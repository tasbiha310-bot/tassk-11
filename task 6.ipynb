```python
import pandas as pd

df = pd.read_csv("coretech_clients_cleaned.csv")

df.head()
```

```python
service_demand = df["service"].value_counts()

print(service_demand)

print(
    "Highest Demand Service:",
    service_demand.idxmax()
)
```

```python
city_clients = df["client_city"].value_counts()

print(city_clients)

print(
    "City with most clients:",
    city_clients.idxmax()
)
```

```python
lead_source = df["lead_source"].value_counts()

print(lead_source)

print(
    "Best Lead Source:",
    lead_source.idxmax()
)
```

```python
avg_budget = df["project_budget"].mean()

print(
    "Average Project Budget:",
    avg_budget
)
```

```python
budget_service = (
    df.groupby("service")["project_budget"]
    .mean()
)

print(budget_service)

print(
    "Highest Budget Service:",
    budget_service.idxmax()
)
```

```python
budget_service = (
    df.groupby("service")["project_budget"]
    .mean()
)

print(budget_service)

print(
    "Highest Budget Service:",
    budget_service.idxmax()
)
```

```python
status_count = df["status"].value_counts()

print(status_count)
```

```python
avg_rating = df["rating"].mean()

print(
    "Average Rating:",
    avg_rating
)
```

```python
print("===== EDA REPORT =====")

print(
"Highest Demand Service:",
df["service"].value_counts().idxmax()
)

print(
"Top City:",
df["client_city"].value_counts().idxmax()
)

print(
"Best Lead Source:",
df["lead_source"].value_counts().idxmax()
)

print(
"Average Budget:",
df["project_budget"].mean()
)

print(
"Highest Budget Service:",
df.groupby("service")["project_budget"].mean().idxmax()
)

print(
"Status Counts:"
)

print(df["status"].value_counts())

print(
"Average Rating:",
df["rating"].mean()
)
```
