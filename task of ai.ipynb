```python
import pandas as pd

df = pd.read_csv("coretech_clients.csv")

df.head()
```

```python
print("Shape:", df.shape)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nData Types:")
print(df.dtypes)
```

```python
# Clean column names
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")

# Handle missing values
df.fillna("Unknown", inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Fix numeric columns
df["project_budget"] = pd.to_numeric(
    df["project_budget"],
    errors="coerce"
)

df["project_duration_days"] = pd.to_numeric(
    df["project_duration_days"],
    errors="coerce"
)

df["rating"] = pd.to_numeric(
    df["rating"],
    errors="coerce"
)
```

```python
print("Shape:", df.shape)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nData Types:")
print(df.dtypes)
```

```python
df.to_csv(
    "coretech_clients_cleaned.csv",
    index=False
)

print("File Saved")
```
